#### checking for outliers within the horsepower column

We will find outliers using boxplot

```python
# importing library

import seaborn as sns
```

```python
sns.boxplot(df['hp'])
```

There is only one outlier within the horsepower column which is greater than 300. Lets find it!

```python
df[df['hp']>300]
```

Maserati Bora model of cars has the maximum hp = 335 which is an outlier! Hence, we will remove it.

```python
df_new = df.drop(30)    # new dataset df_new without outlier
df_new
```

You can remove outlier from the original dataset also without creating new dataset. 

```python
df.drop(df.index[[30]],inplace=True)
df
```
