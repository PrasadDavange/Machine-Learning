#### Identify a few of mostly correlated features, which would help in modification

```python
#importing libraries

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.info()    # info returns the information about the columns of dataset ie., datatype, no. of non-null values,etc
```

```python
df_new = df.drop(['model'],axis=1)   #keeping only numeric columns for computing correlations among them
df_new
```

```python
correlations = df_new.corr()
sns.heatmap(data=correlations,square=True,cmap="bwr")
```
